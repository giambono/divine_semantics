```python
import openai
import pandas as pd

# Set your OpenAI API key
openai.api_key = ""
```

```python
df = pd.read_csv("/content/commedia.csv", delimiter=";", on_bad_lines='skip')
df = df[(df["author"] == "dante") & (df["cantica"] == 1)]

# df = df.iloc[10:20]
df.tail()
```

```python
import time
import pandas as pd

# !openai migrate

# Assuming your paraphrase_text function is defined
def paraphrase_text(text, style="modern simple"):
    prompt = f"Paraphrase the following text in {style} Italian:\n\n{text}"
    response = openai.chat.completions.create(
        model="gpt-3.5-turbo",
        messages=[{"role": "user", "content": prompt}],
        temperature=0.7
    )
    return response.choices[0].message.content


# Load your CSV
# df = pd.read_csv("comedy.csv", delimiter=";")

desired_style = "modern simple"  # or "modern sophisticated"

# Prepare a list to collect paraphrased texts
paraphrased_texts = []

# Define a batch size; adjust based on your API limits and time constraints
batch_size = 100  # for example, process 100 rows at a time

for start in range(0, len(df), batch_size):
    end = start + batch_size
    batch = df.iloc[start:end]
    print(f"Processing rows {start} to {end}")

    # Process each row in the batch
    for text in batch["text"]:
        paraphrased = paraphrase_text(text, style=desired_style)
        paraphrased_texts.append(paraphrased)

    # Optional: Pause between batches to comply with rate limits
    time.sleep(1)

# Add the paraphrased texts as a new column and save to CSV
df["paraphrased_text"] = paraphrased_texts
df["style"] = desired_style
df["author2"] = "gpt-3.5-turbo"
df.to_csv("comedy_paraphrased.csv", sep=";", index=False)

print("Paraphrasing complete. Check 'comedy_paraphrased.csv' for results.")
```
